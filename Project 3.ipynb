```python
#Checking values in column KeyTraits (we saw with the first df that values might need some cleaning)
df['KeyTraits'].value_counts()
```

```python
#Replace the misspelled value to the correct version
df.loc[df['KeyTraits'].str.startswith('Rr'), 'KeyTraits'] = 'Resilience'
```

```python
#Checking missing values
df.isna().sum()
```

```python
#Checking wrong values in City
df['City'].value_counts()
```

```python
#Replace the misspelled value to the correct version
df.loc[df['City'].str.startswith('no'), 'City'] = 'No'
df.loc[df['City'].str.startswith('unk'), 'City'] = np.nan
```

```python
#Replace missing ages by the mean
df.loc[(df['Age'].isnull()), 'Age'] = 20
```

```python
#Replace missing values by NaN
df.loc[(df['MentalDisorder'].isnull()), 'MentalDisorder'] = "no_value"
df.loc[df['MentalDisorder'].str.startswith('no_value'), 'MentalDisorder'] = np.nan
```

```python
#Checking indicator mean value
df.describe()
```

```python
#Creating a function that will replace the missing values by the mean
def fill_indicator(column, mean):

        df.loc[(df[column].isnull()), column] = mean
```

```python
#Creating dictionnary with the pair column : mean
value = {'Perseverance':3,'DesireToTakeInitiative':3,
        'Competitiveness':4, 'SelfReliance':4, 'StrongNeedToAchieve':4,
        'SelfConfidence':3, 'GoodPhysicalHealth':3}
```

```python
#Loop to use the function fill_indicator with values in the dictionnary
for i in value:
    fill_indicator(i, value[i])
```
